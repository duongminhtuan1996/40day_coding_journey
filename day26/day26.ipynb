Tăng, Giảm độ sáng

```python
#Read image
img = cv2.imread('./dataset/1.jpg')
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# Adjust brightness
new_img = img.astype(np.float32) + 50
new_img = np.clip(new_img, 0, 255)
new_img = new_img.astype(np.uint8)
```

```python
new_img_2 = img.astype(np.float32) - 50
new_img_2 = np.clip(new_img_2, 0, 255)
new_img_2 = new_img_2.astype(np.uint8)
```

```python
fig, axes = plt.subplots(2, 2, figsize=(25, 10))

#Display images
axes[0,0].imshow(img_1, cmap='gray')
# axes[0,0].axis('off')
axes[0,0].set_title('Original Image')

axes[0,1].imshow(new_img, cmap='gray')
axes[0,1].axis('off')
axes[0,1].set_title('Brightness Image')

axes[1,0].imshow(new_img_2, cmap='gray')
axes[1,0].axis('off')
axes[1,0].set_title('Darkness Image')

axes[1, 1].axis('off')
#Show the plot
plt.show()
```
