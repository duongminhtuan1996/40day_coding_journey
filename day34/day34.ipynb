cv.filter2D

```python
def gaussian_kernel(size, sigma):
    if size % 2 == 0:
        size = size + 1
    
    max_point = size // 2 #both directions (x,y) maximum cell start point

    min_point = -max_point #both directions (x, y) minimum cell start point

    K = np.zeros((size, size)) #kernel size
    for x in range(min_point, max_point + 1):
        for y in range(min_point, max_point + 1):
            value = (1/(2*np.pi*sigma**2))*np.exp(-((x**2 + y**2)/(2*sigma**2))) #gaussian Filter Forumla Here#
            K[x - min_point, y - min_point] = value

    return K 
kernel = gaussian_kernel(5, 1.4)
img = cv2.imread('./day26/dataset/1.jpg', 0)
img_gaussian = cv2.filter2D(img, -1, kernel)
```

```python
plt.imshow(img_gaussian, cmap='gray')
```

```python
img_org = cv2.imread('./day26/dataset/2.jpg', 0)
```

```python
kernel_1 = gaussian_kernel(5, 2.5)
kernel_2 = gaussian_kernel(5, 5.0)
kernel_3 = gaussian_kernel(5, 10.0)

img_signa1 = cv2.filter2D(img_org, -1, kernel_1)
img_signa2 = cv2.filter2D(img_org, -1, kernel_2)
img_signa3 = cv2.filter2D(img_org, -1, kernel_3)
```

```python
#Create subplots with a larger figure size
fig, axes = plt.subplots(2, 2, figsize=(15, 10))

#Display
axes[0, 0].imshow(img_org, cmap='gray')
axes[0, 0].axis('off')
axes[0, 0].set_title('Origin Image')

axes[0, 1].imshow(img_signa1, cmap='gray')
axes[0, 1].axis('off')
axes[0, 1].set_title('Sigma = 2.5')

axes[1, 0].imshow(img_signa2, cmap='gray')
axes[1, 0].axis('off')
axes[1, 0].set_title('Sigma = 5.0')

axes[1, 1].imshow(img_signa3, cmap='gray')
axes[1, 1].axis('off')
axes[1, 1].set_title('Sigma = 10.0')
```
